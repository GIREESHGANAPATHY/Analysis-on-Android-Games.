## Visualization

```python
sns.set(style="darkgrid")
a=sns.FacetGrid(df,col="installs",height=4,aspect=0.4)
a.map(plt.hist,"Average rating")
```

```python
sns.catplot(x="Average rating",y="installs",col="Paid",kind="strip",data=df,height=8,aspect=.8)
```

```python
sns.catplot(y="Category",hue="installs",kind="count",data=df,height=10,aspect=0.8)
```

```python
sns.catplot(x="Average rating",y="Category",col="installs",kind="box",data=df,height=12,aspect=0.2)
```
